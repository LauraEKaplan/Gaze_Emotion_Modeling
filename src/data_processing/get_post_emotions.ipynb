```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CHANGE PATH IF RUNNING CODE
path = '/Users/sallykim/Desktop/cs136/project/archive/03 Self-Reported Questionnaires'

pre_survey = pd.read_excel(path + "/01 Participant Profile  Pre-Exposure Ratings.xlsx")
post_survey = pd.read_excel(path + "/02 Post Exposure Ratings.xlsx")
```

```python
pre_labels = ["ID",
              "PRE_Joy", "PRE_Happiness", "PRE_Calmness", "PRE_Relaxation", "PRE_Anger", "PRE_Disgust",
              "PRE_Fear", "PRE_Anxiousness", "PRE_Sadness"]

pre_emotions = pre_survey[pre_labels]

pre_emotions.columns = ["ID", "Joy", "Happiness", "Calmness", "Relaxation", "Anger", "Disgust", "Fear",
                        "Anxiousness", "Sadness"]
```

```python
post_labels = ["ID", "Str_Code", "Quad_Cat",
               "POST_Joy", "POST_Happy", "POST_Calmness", "POST_Relaxation", "POST_Anger", "POST_Disgust",
               "POST_Fear", "POST_Anxiousness", "POST_Sadness"]

post_emotions = post_survey[post_labels]

post_emotions.columns = ["ID", "Str_Code", "Quad_Cat", "Joy", "Happiness", "Calmness", "Relaxation", "Anger",
                       "Disgust", "Fear", "Anxiousness", "Sadness"]
```

```python
pre = pre_emotions.set_index(["ID"])
post = post_emotions.set_index(["ID", "Str_Code", "Quad_Cat"])
```

```python
pre.sort_index().loc[101]
```

```python
post.sort_index().loc[101]
```

```python
post_norm = post.sub(pre, axis=0).div(100, axis=0)
```
